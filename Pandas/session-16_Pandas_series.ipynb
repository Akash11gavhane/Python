### **`Series from Dictionary`**

        here we make series using dictionary

```python
marks = {

    "maths" : 67 , 
    "english" : 57,
    "science" : 89,
    "marathi" : 100

}

marks_series = pd.Series(marks , name = "Akash marks")
marks_series
```

### **`Series Method`**

```python
# head and tail

subs.head()

# head function is used to see the top entries in the dataset
# we can pass the number inside the function 
# the number that how many entries we want to see

# eg.

# by default value is 5

print(subs.head(10))
print(subs.head(3))
print(subs.head(30))
```

```python
# tail 

# tail function is used to get last 5 rows 

kohli_ipl.tail()

# we can get custom number of rows by passing the number of rows that you want 

print(kohli_ipl.tail(10))
print(kohli_ipl.tail(15))
print(kohli_ipl.tail(3))
```

```python
# sample()

# sample function randomely display one row from the dataset

movies.sample()

# we can get custom random movies also
# default parameter is 1

print(movies.sample(5))
print("-------------------------------------------------------")
print(movies.sample(10))
print("-------------------------------------------------------")
print(movies.sample(15))
```

```python
# value_counts()

# value_counts - it give the frequancy of every item in the data set.

movies.value_counts()


# here we get which actor do work in how many movies
```

```python
# sort_values

# it sort all the values

kohli_ipl.sort_values()
```

```python
kohli_ipl.sort_values(ascending=False)

# to get the values in the reverse order or descending order
```

```python
# get only highest runs

kohli_ipl.sort_values(ascending=False).head(1)
```

```python
kohli_ipl.sort_values(ascending=False).head(1).values[0]

# this is called as method chaining ---> we use function over function
# here the first method ouput is input for his nexg method and so on.
```

```python
kohli_ipl.sort_values(inplace = True)

# using that inplace parameter we can change in the original dataset

# not in the original file 
# in the variable where we stored that data
```

```python
# sort_index

# this function is used to sort the index

movies.sort_index()

# sort movies in ascending order
```

```python
movies.sort_index(ascending= False)
```

```python
movies.sort_index(inplace = True)

# here also we have inplace parameter

```
