```python
import numpy as np

import pandas as pd
```

### **`Creating DataFrames`**

```python
# using lists
# create dataframe using list

# if you want to make a dataframe then we need a 2D list --> list inside list

student_data = [
    [100 , 80 , 10],
    [90,70,7],
    [120,100,14],
    [80,50,2]
]

pd.DataFrame(student_data , columns = ['iq' , 'marks' , 'package'])

# here using colummn parameter we specify the names for our column
```

```python
# using dictionary

# create a dataframe using dictionary

student_dict = {
    'iq' : [100 , 90 , 120 , 80],
    'marks' : [80 , 70 , 100 , 50],
    'package' : [10 , 7 , 14 , 2]
}

student = pd.DataFrame(student_dict)

student

# here there is no need to specify the column names like we specify in list
```

```python
student_dict = {
    'name':['nitish','ankit','rupesh' , 'Akash'],
    'iq' : [100 , 90 , 120 , 80],
    'marks' : [80 , 70 , 100 , 50],
    'package' : [10 , 7 , 14 , 8]
}

student = pd.DataFrame(student_dict)

student
```
