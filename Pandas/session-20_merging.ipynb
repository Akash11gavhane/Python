### Concat Two or more DataFrame or Datasets

```python
# pd.concat
# ignore_index
# df.append
# mullitindex -> fetch using iloc
# concat dataframes horizontally
```

```python
#  pd.concat

pd.concat([nov , dec]) # we have to pass the name of datasets in to the list

# concat function is used to concat or merge the two data set vertically

# here using concat function we merge the datasets

# problem is here indexes are repeat

# to solve that problem we have parameter called

# index_parameter = True
```

```python
pd.concat([nov , dec] , ignore_index=True)

# now here the index are not repeated
```

```python
regs = pd.concat([nov , dec] , ignore_index=True)

regs.head()
```

```python
# how can we do same work like above (concat)
# we have another method for that


nov.append(dec)

# use concat always 
# append is going to depricate from upcoming pandas version

# append is work same like concat
```

```python
# mullitindex -> fetch using iloc

multi = pd.concat([nov , dec] , keys = ['November' , 'December'])

# we get a multiindex data frame 
# above we have a problem about index repetation 
# but here we can keep that indexes
# using keys = [] parameter
# where we have to pass new index names for the perticular data in order

# multiindex dataframe means we have multiple index

# now we get data 
# which have november index for nov data
# and december index for dec data 
```

```python
# how to work with multiindex dataset

# extract only november registrations

multi.loc['November']

# here we get only data of november month
```

```python
# extract the 1st person from the november month 

multi.loc[('November' , 0)]

```

```python
# # extract the 5th person from the december month 

multi.loc[("December" , 4)]
```

```python
# how to concat or merge the data side by side

pd.concat([nov , dec] , axis = 1)
```

```python
# inner Join

# we have two table 

# 1st one is left table
# 2nd one is right table

# if we have to join them we find the items which are present in both tables

# if we perform inner join then we find those element which are present in both tables means we find common elements

# if we performing inner join means we are finding common elements.

students
```

```python
# there are 25 students
# and regs have 53 students means courses are brought 53 times

# so here we perform inner join on :

# students
# and 
# regs

# join always perform using common column 

# so here student_id is common column

# so here we can see only those students who are in the both students and regs 

```

```python
# performing inner join

# for that we use merge method 

students

# which is comes first that become our left table and second one become right table


students.merge(regs , how = 'inner' , on = 'student_id')

# here 
# how = 'inner' means which join operation we have to perform

# and 

# on = 'column_name' --> here in on we have to give the name of column on which we are going to perform operation.
```
